```python
#fUNCION para aplicar el diccionario de datos
def aplicar_diccionarios(df, diccionario_datos, diccionario_datos_inv):
    for col, dic in diccionario_datos.items():
        if col in df.columns:
            df[col] = df[col].map(dic)
    
    # for para pegar el ID 
    for col, dic in diccionario_datos_inv.items():
        if col in df.columns:
            df[col] = df[col].map(dic)
    
    return df
```
